```python
import pandas as pd
import numpy as np
import json
from glob import glob
import matplotlib.pyplot as plt
```

```python
lstm_mae = pd.read_csv("./best_models/lstm/run-17-04-2023_231059_tensorboard-tag-Training MAE per Epoch.csv")
cnn_lstm_mae = pd.read_csv("./best_models/cnn-lstm/run-17-04-2023_190141_tensorboard-tag-Training MAE per Epoch.csv")
transformer_mae = pd.read_csv("./best_models/transformer/run-17-04-2023_214336-tag-Training MAE per Epoch.csv")
cnn_transformer_mae = pd.read_csv("./best_models/cnn-transformer/run-17-04-2023_160328-tag-Training MAE per Epoch.csv")

lstm_mse = pd.read_csv("./best_models/lstm/run-17-04-2023_231059_tensorboard-tag-Training MSE per Epoch.csv")
cnn_lstm_mse = pd.read_csv("./best_models/cnn-lstm/run-17-04-2023_190141_tensorboard-tag-Training MSE per Epoch.csv")
transformer_mse = pd.read_csv("./best_models/transformer/run-17-04-2023_214336-tag-Training MSE per Epoch.csv")
cnn_transformer_mse = pd.read_csv("./best_models/cnn-transformer/run-17-04-2023_160328-tag-Training MSE per Epoch.csv")


val_lstm_mae = pd.read_csv("./best_models/lstm/run-17-04-2023_231059_tensorboard-tag-Validation MAE per Epoch.csv")
val_cnn_lstm_mae = pd.read_csv("./best_models/cnn-lstm/run-17-04-2023_190141_tensorboard-tag-Validation MAE per Epoch.csv")
val_transformer_mae = pd.read_csv("./best_models/transformer/run-17-04-2023_214336-tag-Validation MAE per Epoch.csv")
val_cnn_transformer_mae = pd.read_csv("./best_models/cnn-transformer/run-17-04-2023_160328-tag-Validation MAE per Epoch.csv")

val_lstm_mse = pd.read_csv("./best_models/lstm/run-17-04-2023_231059_tensorboard-tag-Validation MSE per Epoch.csv")
val_cnn_lstm_mse = pd.read_csv("./best_models/cnn-lstm/run-17-04-2023_190141_tensorboard-tag-Validation MSE per Epoch.csv")
val_transformer_mse = pd.read_csv("./best_models/transformer/run-17-04-2023_214336-tag-Validation MSE per Epoch.csv")
val_cnn_transformer_mse = pd.read_csv("./best_models/cnn-transformer/run-17-04-2023_160328-tag-Validation MSE per Epoch.csv")
```

```python
print("MAE")
print(f"\t LSTM: {round(val_lstm_mae['Value'].min(), 5)}")
print(f"\t CNN-LSTM: {round(val_cnn_lstm_mae['Value'].min(), 5)}")
print(f"\t Transformer: {round(val_transformer_mae['Value'].min(), 5)}")
print(f"\t CNN-Transformer: {round(val_cnn_transformer_mae['Value'].min(), 5)}")
print("MSE")
print(f"\t LSTM: {round(val_lstm_mse['Value'].min(), 6)}")
print(f"\t CNN-LSTM: {round(val_cnn_lstm_mse['Value'].min(), 6)}")
print(f"\t Transformer: {round(val_transformer_mse['Value'].min(), 6)}")
print(f"\t CNN-Transformer: {round(val_cnn_transformer_mse['Value'].min(), 6)}")
```

```python
fig, ax = plt.subplots()

ax.set_ylabel("MAE loss")
ax.set_xlabel("Epoch")


ax.plot(lstm_mae['Step'], lstm_mae['Value'], label="LSTM")
ax.plot(cnn_lstm_mae['Step'], cnn_lstm_mae['Value'], label="CNN-LSTM")
ax.plot(transformer_mae['Step'], transformer_mae['Value'], label="Transformer")
ax.plot(cnn_transformer_mae['Step'], cnn_transformer_mae['Value'], label="CNN-Transformer")

plt.legend()
plt.grid('on')
plt.tight_layout()
plt.title("Training MAE per Epoch")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_ylabel("MSE loss")
ax.set_xlabel("Epoch")


ax.plot(lstm_mse['Step'], lstm_mse['Value'], label="LSTM")
ax.plot(cnn_lstm_mse['Step'], cnn_lstm_mse['Value'], label="CNN-LSTM")
ax.plot(transformer_mse['Step'], transformer_mse['Value'], label="Transformer")
ax.plot(cnn_transformer_mse['Step'], cnn_transformer_mse['Value'], label="CNN-Transformer")

plt.legend()
plt.grid('on')
plt.tight_layout()
plt.title("Training MSE per Epoch")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_ylabel("MAE loss")
ax.set_xlabel("Epoch")


ax.plot(val_lstm_mae['Step'], val_lstm_mae['Value'], label="LSTM")
ax.plot(val_cnn_lstm_mae['Step'], val_cnn_lstm_mae['Value'], label="CNN-LSTM")
ax.plot(val_transformer_mae['Step'], val_transformer_mae['Value'], label="Transformer")
ax.plot(val_cnn_transformer_mae['Step'], val_cnn_transformer_mae['Value'], label="CNN-Transformer")

plt.legend()
plt.grid('on')
plt.tight_layout()
plt.title("Validation MAE per Epoch")

plt.show()
```
